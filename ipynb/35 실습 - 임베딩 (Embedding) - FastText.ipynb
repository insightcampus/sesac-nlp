#FastText

```
import pandas as pd
it_df = pd.read_csv('./word_embedding_it.csv').dropna()
culture_df = pd.read_csv('./word_embedding_culture.csv').dropna()

it_token_ls = list(it_df.loc[:,'1']) 
culture_token_ls = list(culture_df.loc[:,'1'])

total_token_ls = it_token_ls + culture_token_ls  

# 리스트의 값을 ','로 분리
total_token_ls = [tokens.split(',') for tokens in total_token_ls]
```

```
total_token_ls[:50]
```

```
from gensim.models import FastText

# FastText 모델생성
ft_model = FastText(total_token_ls,  
                      size=10,                 # 임베딩된 단어 벡터의 차원 크기
                      window=8,                 # 문맥의 크기(window_size)
                      min_count=2,            # 2번 미만 등장한 단어는 제외
                      alpha = 0.025,                    # 학습률(Learning rate)
                      sg = 1,                             # 0: CBOW, 1: Skip-gram
                      iter = 10,              # 전체 문장 반복학습 횟수(epoch)
                      min_n=3, max_n=6)            # 최소, 최대 N-gram 수   
```

```
ft_model.wv['인공']
```

```
ft_model.wv['인공지능']
```

```
# Getting most similar vectors
print(ft_model.wv.most_similar('인공'))

# Getting most similar vectors
print(ft_model.wv.most_similar('인공지능'))
```
